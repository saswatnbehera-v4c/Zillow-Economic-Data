## Starting Cleanup of Previous Runs

```python
target_dir = output_dir_path 

print(f"--- Starting Cleanup in: {target_dir} ---")

if os.path.exists(target_dir):
    files = os.listdir(target_dir)
    deleted_count = 0
    
    for filename in files:
        if "chunk" in filename:
            full_path = os.path.join(target_dir, filename)
            try:
                os.remove(full_path)
                print(f"Deleted: {filename}")
                deleted_count += 1
            except Exception as e:
                print(f"⚠️ Could not delete {filename}: {e}")
                
    print(f"--- Cleanup Complete. Removed {deleted_count} files. ---")
else:
    print(f"Target directory {target_dir} does not exist. Nothing to clean.")
```
